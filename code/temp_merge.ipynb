```python
import pandas as pd
from tqdm import tqdm
```

```python
df_1 = pd.read_csv('./result/patent_sim_1.csv')
df_1.head()
```

```python
df_2 = pd.read_csv('./result/patent_sim_2.csv')
df_2.head()
```

```python
df_3 = pd.read_csv('./result/patent_sim_3.csv')
df_3.head()
```

#### Step 合并

```python
merged_df = pd.concat([df_1, df_2, df_3], axis=0, ignore_index=True)
```

```python
merged_df
```

```python
df_unique = merged_df.drop_duplicates()
```

```python
df_unique.to_csv('./result/p2p_reference.sim.csv',index=False)
```

```python
summary = df_unique.groupby('category')['similarity'].agg(['mean', 'count', 'std']).reset_index()
print(summary)
```
