```python
import os
import pandas as pd

# 指定TSV文件路径
file_path = '../g_inventor_disambiguated.tsv/g_inventor_disambiguated.tsv'  # 替换为你的TSV文件路径

# 检查文件是否存在
if not os.path.exists(file_path):
    print(f"文件不存在：{file_path}")
else:
    print(f"文件存在：{file_path}")

    # 检查文件权限
    if os.access(file_path, os.R_OK):
        print(f"文件可读：{file_path}")
    else:
        print(f"文件不可读，正在修改权限...")
        try:
            # 修改文件权限（仅在Linux/macOS中有效）
            os.chmod(file_path, 0o644)
            print(f"文件权限已修改")
        except PermissionError as e:
            print(f"无法修改文件权限：{e}")

# 尝试读取TSV文件
try:
    df = pd.read_csv(file_path, sep='\t')
    print(df.head())
except PermissionError as e:
    print(f"权限错误：{e}")
except Exception as e:
    print(f"读取文件时发生错误：{e}")
```
